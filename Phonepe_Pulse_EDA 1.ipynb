# **1. Aggregated Data -**

### 1.a) Aggregated Transaction

````python
data_list = []
path = "/content/pulse/data/aggregated/transaction/country/india/state"

for state in sorted(os.listdir(path)):
    spath = os.path.join(path, state)

    for year in sorted(os.listdir(spath)):
        ypath = os.path.join(spath, year)

        for file in sorted(os.listdir(ypath)):
            if file.endswith(".json"):
                with open(os.path.join(ypath, file)) as f:
                    data = json.load(f)

                try:
                    for item in data["data"]["transactionData"]:
                        for p in item["paymentInstruments"]:
                            data_list.append({
                                "State": state,
                                "Year": int(year),
                                "Quarter": int(file.replace(".json","")),
                                "Category": item["name"],
                                "Count": p["count"],
                                "Amount": p["amount"]
                            })
                except:
                    continue

agg_transaction = pd.DataFrame(data_list).sort_values(["Year","Quarter"]).reset_index(drop=True)

print("Years:", sorted(agg_transaction["Year"].unique()))

agg_transaction

agg_transaction.groupby(["Year","State","Category"])[["Count","Amount"]].sum().reset_index()
````

### 1.b) Aggregated User

````python
data_list = []
path = "/content/pulse/data/aggregated/user/country/india/state"

for state in sorted(os.listdir(path)):
    spath = os.path.join(path, state)

    for year in sorted(os.listdir(spath)):
        ypath = os.path.join(spath, year)

        for file in sorted(os.listdir(ypath)):
            if file.endswith(".json"):
                with open(os.path.join(ypath, file)) as f:
                    data = json.load(f)

                try:
                    for item in data["data"]["usersByDevice"]:
                        data_list.append({
                            "State": state,
                            "Year": int(year),
                            "Quarter": int(file.replace(".json","")),
                            "Brand": item["brand"],
                            "Count": item["count"],
                            "Percentage": item["percentage"]
                        })
                except:
                    continue

agg_user = pd.DataFrame(data_list).sort_values(["Year","Quarter"]).reset_index(drop=True)

print("Years:", sorted(agg_user["Year"].unique()))

agg_user

agg_user.groupby(["Year","State","Brand"])[["Count"]].sum().reset_index()
````

### 1C. Aggregated Insurance

````python
data_list = []
path = "/content/pulse/data/aggregated/insurance/country/india/state"

for state in sorted(os.listdir(path)):
    spath = os.path.join(path, state)

    for year in sorted(os.listdir(spath)):
        ypath = os.path.join(spath, year)

        for file in sorted(os.listdir(ypath)):
            if file.endswith(".json"):
                with open(os.path.join(ypath, file)) as f:
                    data = json.load(f)

                try:
                    for item in data["data"]["transactionData"]:
                        for p in item["paymentInstruments"]:
                            data_list.append({
                                "State": state,
                                "Year": int(year),
                                "Quarter": int(file.replace(".json","")),
                                "Category": item["name"],
                                "Count": p["count"],
                                "Amount": p["amount"]
                            })
                except:
                    continue

agg_insurance = pd.DataFrame(data_list).sort_values(["Year","Quarter"]).reset_index(drop=True)

print("Years:", sorted(agg_insurance["Year"].unique()))

agg_insurance

agg_insurance.groupby(["Year","State","Category"])[["Count","Amount"]].sum().reset_index()
````

# 2. Map Data -

### 2.a) Map Transaction

````python
data_list = []
path = "/content/pulse/data/map/transaction/hover/country/india"

for year in sorted(os.listdir(path)):
    ypath = os.path.join(path, year)

    for file in sorted(os.listdir(ypath)):
        if file.endswith(".json"):
            with open(os.path.join(ypath, file)) as f:
                data = json.load(f)

            try:
                for item in data["data"]["hoverDataList"]:
                    data_list.append({
                        "Year": int(year),
                        "Quarter": int(file.replace(".json","")),
                        "State": item["name"],
                        "Count": item["metric"][0]["count"],
                        "Amount": item["metric"][0]["amount"]
                    })
            except:
                continue

map_transaction = pd.DataFrame(data_list)
map_transaction
````

### 2.b) Map User

````python
data_list = []
path = "/content/pulse/data/map/user/hover/country/india"

for year in sorted(os.listdir(path)):
    ypath = os.path.join(path, year)

    for file in sorted(os.listdir(ypath)):
        if file.endswith(".json"):
            with open(os.path.join(ypath, file)) as f:
                data = json.load(f)

            try:
                for state, val in data["data"]["hoverData"].items():
                    data_list.append({
                        "Year": int(year),
                        "Quarter": int(file.replace(".json","")),
                        "State": state,
                        "RegisteredUsers": val["registeredUsers"],
                        "AppOpens": val["appOpens"]
                    })
            except:
                continue

map_user = pd.DataFrame(data_list)
map_user
````

### 2.c) Map Insurance

````python
data_list = []
path = "/content/pulse/data/map/insurance/hover/country/india"

for year in sorted(os.listdir(path)):
    ypath = os.path.join(path, year)

    for file in sorted(os.listdir(ypath)):
        if file.endswith(".json"):
            with open(os.path.join(ypath, file)) as f:
                data = json.load(f)

            try:
                for item in data["data"]["hoverDataList"]:
                    data_list.append({
                        "Year": int(year),
                        "Quarter": int(file.replace(".json","")),
                        "State": item["name"],
                        "Count": item["metric"][0]["count"],
                        "Amount": item["metric"][0]["amount"]
                    })
            except:
                continue

map_insurance = pd.DataFrame(data_list)
map_insurance
````

# 3. Top Data -

### 3.a) Top Transaction

````python
data_list = []

path = "/content/pulse/data/top/transaction/country/india/state/"

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    for year in sorted(os.listdir(state_path)):
        year_path = os.path.join(state_path, year)

        for file in sorted(os.listdir(year_path)):
            if file.endswith(".json"):
                file_path = os.path.join(year_path, file)

                with open(file_path) as f:
                    data = json.load(f)

                try:
                    districts = data["data"]["districts"]

                    for item in districts:
                        data_list.append({
                            "Year": int(year),
                            "Quarter": int(file.replace(".json","")),
                            "State": state.replace("-", " ").title(),
                            "Count": item["metric"]["count"],
                            "Amount": item["metric"]["amount"]
                        })

                except:
                    continue

# Create DataFrame
top_transaction = pd.DataFrame(data_list)\
    .sort_values(["Year","Quarter"])\
    .reset_index(drop=True)

# Check years
print("Years:", sorted(top_transaction["Year"].unique()))

top_transaction
````

### 3.b) Top User

````python
data = []

path = "pulse/data/top/user/country/india/state/"

for state in os.listdir(path):
    for year in sorted(os.listdir(path + state)):
        for file in sorted(os.listdir(path + state + "/" + year)):
            with open(path + state + "/" + year + "/" + file) as f:
                d = json.load(f)
                for item in d["data"]["districts"]:
                    data.append({
                        "Year": int(year),
                        "Quarter": int(file.replace(".json","")),
                        "State": state.replace("-", " ").title(),
                        "RegisteredUsers": item["registeredUsers"]
                    })

top_user = pd.DataFrame(data).sort_values(["Year","Quarter"]).reset_index(drop=True)

print("Years:", sorted(top_user["Year"].unique()))
top_user
````

### **3.c) Top Insurance**

````python
data = []

path = "pulse/data/top/insurance/country/india/state/"

for state in os.listdir(path):
    for year in sorted(os.listdir(path + state)):
        for file in sorted(os.listdir(path + state + "/" + year)):
            with open(path + state + "/" + year + "/" + file) as f:
                d = json.load(f)
                for item in d["data"]["districts"]:
                    data.append({
                        "Year": int(year),
                        "Quarter": int(file.replace(".json","")),
                        "State": state.replace("-", " ").title(),
                        "Count": item["metric"]["count"],
                        "Amount": item["metric"]["amount"]
                    })

top_insurance = pd.DataFrame(data).sort_values(["Year","Quarter"]).reset_index(drop=True)

print("Years:", sorted(top_insurance["Year"].unique()))
top_insurance
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling Code (SQL-compatible & final)

# 🔹 Step 1: List all datasets
datasets = [
    agg_transaction, agg_user, agg_insurance,
    map_transaction, map_user, map_insurance,
    top_transaction, top_user, top_insurance
]

# 🔹 Step 2: Clean each dataset properly
for i, df in enumerate(datasets):

    # Standardize column names
    df.columns = df.columns.str.strip()

    # Fix data types
    if "Year" in df.columns:
        df["Year"] = df["Year"].astype(int)
    if "Quarter" in df.columns:
        df["Quarter"] = df["Quarter"].astype(int)

    # Clean state names
    if "State" in df.columns:
        df["State"] = df["State"].str.replace("-", " ").str.title()

    # Remove duplicates
    df = df.drop_duplicates()

    # Save back to list
    datasets[i] = df


# 🔹 Step 3: Reassign cleaned data back to original variables
(
    agg_transaction, agg_user, agg_insurance,
    map_transaction, map_user, map_insurance,
    top_transaction, top_user, top_insurance
) = datasets


# 🔹 Step 4: Save CLEANED data back to SQL (VERY IMPORTANT)

agg_transaction.to_sql("agg_transaction", engine, if_exists="replace", index=False)
agg_user.to_sql("agg_user", engine, if_exists="replace", index=False)
agg_insurance.to_sql("agg_insurance", engine, if_exists="replace", index=False)

map_transaction.to_sql("map_transaction", engine, if_exists="replace", index=False)
map_user.to_sql("map_user", engine, if_exists="replace", index=False)
map_insurance.to_sql("map_insurance", engine, if_exists="replace", index=False)

top_transaction.to_sql("top_transaction", engine, if_exists="replace", index=False)
top_user.to_sql("top_user", engine, if_exists="replace", index=False)
top_insurance.to_sql("top_insurance", engine, if_exists="replace", index=False)


# 🔹 Step 5: Verify shapes
print("🔹 Dataset Shapes After Cleaning:")

print("Agg Transaction:", agg_transaction.shape)
print("Agg User:", agg_user.shape)
print("Agg Insurance:", agg_insurance.shape)

print("Map Transaction:", map_transaction.shape)
print("Map User:", map_user.shape)
print("Map Insurance:", map_insurance.shape)

print("Top Transaction:", top_transaction.shape)
print("Top User:", top_user.shape)
print("Top Insurance:", top_insurance.shape)
````

In the data wrangling process, several important steps were performed to ensure that the dataset is clean, consistent, and ready for analysis.

First, column names were standardized by removing any unwanted spaces to maintain uniformity across all datasets. Then, the data types of key columns such as Year and Quarter were converted into integer format to ensure proper time-based analysis and avoid inconsistencies.

#### Chart - 14 - Correlation Heatmap

````python
# Chart - 14 : Correlation Heatmap (SQL-based)

query = """
SELECT
    a.Year,
    SUM(a.Amount) AS Total_Amount,
    SUM(a.Count) AS Total_Count,
    SUM(u.RegisteredUsers) AS Total_Users,
    SUM(u.AppOpens) AS Total_AppOpens
FROM agg_transaction a
JOIN map_user u
ON a.Year = u.Year
GROUP BY a.Year
ORDER BY a.Year
"""

heatmap_data = pd.read_sql(query, engine)

# Drop Year column for correlation
corr_matrix = heatmap_data.drop(columns=["Year"]).corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Key Metrics")

plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used to analyze relationships between multiple numerical variables simultaneously. It helps in identifying how strongly variables are related to each other, whether positively or negatively. This chart provides a clear overview of interdependencies among key metrics such as transaction amount, transaction count, registered users, and app opens.


##### 2. What is/are the insight(s) found from the chart?

The heatmap shows strong positive correlations among all key variables. Transaction amount and transaction count are highly correlated, indicating that higher activity leads to higher value. Similarly, registered users and app opens also show a strong positive relationship, reflecting that increased user base results in higher engagement. Overall, the metrics move together, indicating consistent growth across the platform.


#### Chart - 15 - Pair Plot

````python
# Chart - 15 : Pair Plot (SQL-based)

query = """
SELECT
    a.Year,
    SUM(a.Amount) AS Total_Amount,
    SUM(a.Count) AS Total_Count,
    SUM(u.RegisteredUsers) AS Total_Users,
    SUM(u.AppOpens) AS Total_AppOpens
FROM agg_transaction a
JOIN map_user u
ON a.Year = u.Year
GROUP BY a.Year
ORDER BY a.Year
"""

pairplot_data = pd.read_sql(query, engine)

# Drop Year for pairplot
pairplot_data = pairplot_data.drop(columns=["Year"])

sns.pairplot(pairplot_data, diag_kind="kde")

plt.suptitle("Pair Plot of Key Metrics", y=1.02)

plt.show()
````
